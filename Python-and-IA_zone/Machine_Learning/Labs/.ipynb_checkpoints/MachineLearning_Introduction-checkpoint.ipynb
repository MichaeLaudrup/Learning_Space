```python
%matplotlib notebook
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split


frutas = pd.read_table('data/fruit_data_with_colors.txt')
frutas.head(10)
```

## Visualización de la información

##  Create train-test split
Esta función nos sirve para separar el conjunto de datos inicial en dos subconjuntos, uno de entrenamiento y otro de test

```python
from matplotlib import colormaps

X = frutas[['mass', 'width', 'height', 'color_score']]
y = frutas['fruit_label']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

cmap = colormaps['gnuplot']
scatter = pd.plotting.scatter_matrix(X_train, c=y_train, marker='o', s=40, hist_kwds={'bins':15}, figsize=(8,8), cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'], c=y_train, marker='o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```
