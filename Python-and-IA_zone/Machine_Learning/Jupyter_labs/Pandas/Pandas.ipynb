## Loc
Loc es para buscar por etiquetas de indice, es decir, por los nombres de los campos, o por objetos complejos en caso de que este indexado por objetos complejos

```python
precio_coches = pd.Series({
    'seat_ibiza': 2500,
    'seat_leon': 3100,
    'toyota': 4356,
    'peugeout': 3394,
    'ford': 3423,
    'corolla': 2343,
    'ferrari': 3213213
})

print(f'precio_coches:\n ${precio_coches}')
print(f'precio_coches.loc["seat_ibiza"]: \n{precio_coches["seat_ibiza"]}')
print(f'precio_coches.loc["seat_ibiza":"toyota"]: \n{precio_coches["seat_ibiza":"toyota"]}')
```

# Missing values en Pandas
Cuando transformamos un archivo csv a un dataframe hay muchos valores que no se entiende,en pandas, estos valores, dependiendo del tipo de implicito en cada serie serán Null, Nan, etc. Por ejemplo,

* Si la serie de la columna consiste en números de coma flotante, los valores faltantes se representan como NaN (Not a Number, o "No es un número").
* Si la serie de la columna consiste en enteros, los valores faltantes se representan como NaN. Es importante tener en cuenta que esto es aplicable a partir de la versión 1.0 de Pandas; en versiones anteriores, los arreglos de enteros no podían contener valores NaN.
* Si la serie de la columna consiste en cadenas de texto, los valores faltantes se representan como None.
* Si la serie de la columna consiste en valores booleanos, los valores faltantes se representan como None o NaN.
* Si la serie de la columna consiste en objetos de Python o tipos mixtos, los valores faltantes generalmente se representan como None.
* Si la serie de la columna consiste en objetos de tipo datetime, los valores faltantes se representan como NaT (Not a Time, o "No es una hora").

Cada una de estas representaciones tiene sus propias implicaciones y métodos para manejarlas. Puedes utilizar funciones de Pandas como isna(), fillna() y dropna() para detectar y manejar eficazmente los valores faltantes.

```python
import pandas as pd
df = pd.read_csv('data/class_grades.csv')
df.head()
```

```python
# devuelve a true valores Nan, NoT, None, N/A
mascara_valores_nulos = df.isnull()
mascara_valores_nulos.head(10)
```

```python
# rellena valores nulos con
df.fillna(0, inplace=True)
df.head(10)
```

## Manipulando un dataFrame en panda

```python
import pandas as pd
df = pd.read_csv("data/presidents.csv")
df.head()
```

```python
df["First"] = df['President'] # Creamos nueva columna a partir de otra
df["First"] = df['First'].replace("[ ].*", "", regex=True)

df.head()
```

```python
def splitname(row):
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    return row
df = df.apply(splitname, axis='columns')
df.head(10)
```
