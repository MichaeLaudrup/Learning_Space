## Apply function (Importante para IA)

La función apply en un DataFrame de la biblioteca pandas en Python se utiliza para aplicar una función a lo largo de una fila o columna del DataFrame. Permite realizar operaciones personalizadas en los datos del DataFrame, ya sea en una fila específica o en una columna completa.

La sintaxis básica de apply es la siguiente:

    df.apply(func, axis=0)

Donde:

    func: La función que deseas aplicar a cada fila o columna.
    axis: Indica si deseas aplicar la función a lo largo de las columnas (axis=0) o a lo largo de las filas (axis=1).

La función apply itera a través de cada elemento de la fila o columna especificada y aplica la función que hayas proporcionado. Puedes utilizar funciones predefinidas de Python o incluso funciones personalizadas definidas por el usuario.

Por ejemplo, si tienes un DataFrame df y deseas calcular la suma de los valores en cada columna, puedes usar apply de la siguiente manera:

```python
import pandas as pd
import numpy as np

# Crear un diccionario con columnas y números aleatorios
data = {
    'Camisas vendidas': np.random.randint(0, 100, size=10),
    'Pantalones': np.random.randint(0, 100, size=10),
    'Calzoncillos': np.random.randint(0, 100, size=10),
    'Zapatos': np.random.randint(0, 100, size=10)
}
# Crear el DataFrame
mydf = pd.DataFrame(data)

def max_min(row):
    dict = {
        'total' : row['Camisas vendidas' : 'Zapatos'].sum(),
        'max' : row.max(),
        'min': row.min(),
        'meam': row.mean()
    }
    
    for key in dict:
        row[key] = dict[key]
    return row


mydf = mydf.apply(max_min, axis=1)
mydf
```
